```python
# initializing the list
random_list = Quarterbackdf["Player"]
frequency = {}

# iterating over the list
for item in random_list:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1

# printing the frequency
frequencydf = pd.DataFrame.from_dict(frequency, orient="index")
print(frequencydf)
```

```python
frequencydf.sort_values(by=0, ascending=False)
```

```python
def unique(list1):
 
    # initialize a null list
    unique_list = []
 
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print (x)
```

```python
tom = Quarterbackdf[Quarterbackdf["Player"] == "Tom Brady"]
tom = tom.tail(5)
tomstat = tom.groupby(Quarterbackdf['Player']).sum()
```

```python
pat = Quarterbackdf[Quarterbackdf["Player"] == "Patrick Mahomes"]
pat = pat.tail(5)
patstat = pat.groupby(Quarterbackdf['Player']).sum()
```

```python
justin = Quarterbackdf[Quarterbackdf["Player"] == "Justin Herbert"]
justin = justin.head(5)
justinstat = round(justin.groupby(Quarterbackdf['Player']).sum()*5/3,1)
```

```python
Peyton = Quarterbackdf[Quarterbackdf["Player"] == "Peyton Manning"]
Peyton = Peyton.tail(5)
peytonstat = Peyton.groupby(Quarterbackdf['Player']).sum()
```

```python
josh = Quarterbackdf[Quarterbackdf["Player"] == "Josh Allen"]
josh = josh.head(4)
joshstat = round(josh.groupby(Quarterbackdf['Player']).sum()*5/4,1)
```

```python
rodgers = Quarterbackdf[Quarterbackdf["Player"] == "Aaron Rodgers"]
rodgers = rodgers.tail(5)
rodgersstat = rodgers.groupby(Quarterbackdf['Player']).sum()
```

```python
top = pd.concat([tomstat,patstat,justinstat,peytonstat, joshstat, rodgersstat])
superbowls = ['3','2','0','0','0', '1']
thestats = top[["Pass Yds","Att", "Cmp","Cmp %", "TD", "INT", "Rate", "1st"]]
thestats["SB"]= superbowls
thestats
```

```python
thestats2 = top[["Cmp %", "TD", "INT", "Rate"]]
thestats2
```

```python
Plotdf = thestats2.swapaxes("index", 'columns')
Plotdf.plot(kind="bar", figsize=(8,8), title="Difference in Top QB's first few seasons")
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

#makes an empty list
tweets_list2=[]
#define the inputs we want to look at for our query
query = '''(GOAT AND quarterback) until:2023-03-12 since:2020-01-01'''
#using a for loop we want to grab tweets based on our query
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(query=query).get_items()): #declare a username 
    # We identify the number of tweets or max number of tweets
    if i>5000: 
        break
    # the list is appending the information from each tweet we grab
    tweets_list2.append([tweet.date, tweet.content, tweet.retweetCount, tweet.user.username]) #declare the attributes to be returned
    if i == 500:
        print(dir(tweet))
# Creating a dataframe from the tweets list above 
tweets_df = pd.DataFrame(tweets_list2, columns=['Datetime','Text', "RetweetedCount", "Team"])
tweets_df
```

```python
words1 = tweets_df["Text"]
words1
```

```python
import nltk
import matplotlib.pyplot as plt
```

```python
from nltk.tokenize import word_tokenize
word_tokens1 = []
for sent in words1:
    print(word_tokenize(sent))
    word_tokens1.append(word_tokenize(sent))
```

```python
# Creating a word cloud
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from nltk.corpus import stopwords
stop_words = set (stopwords.words( 'english' ))
# remove url language
stop_words.update(["https", "co", "quarterback"])
textt = " ".join(tweet for tweet in words1)
# set minimum word length
wordcloud = WordCloud(stopwords=stop_words, min_word_length = 4).generate(textt)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
